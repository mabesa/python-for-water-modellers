## What is VS Code?

**Visual Studio Code (VS Code)** is a free text editor designed for writing code. Think of it like Microsoft Word, but for programming instead of documents.

### Why VS Code?

| Feature | Benefit for Water Modellers |
|---------|-----------------------------|
| **Free & Open Source** | No license costs, use anywhere |
| **Cross-platform** | Works on Windows, Mac, and Linux |
| **Excellent Python support** | Syntax highlighting, debugging, IntelliSense |
| **Integrated terminal** | Run scripts without switching windows |
| **Extensions** | Add features like Jupyter notebook support |
| **Widely used** | Lots of tutorials and community support |

### Why Not Just Use Notepad or a Simple Text Editor?

You *could* write Python in Notepad, but VS Code offers:

- **Syntax highlighting** â€” Different parts of code appear in different colors, making it easier to read
- **Auto-completion** â€” Suggests code as you type, reducing typos
- **Error detection** â€” Shows problems before you run the code
- **Integrated terminal** â€” Run your scripts without leaving the editor
- **File explorer** â€” Navigate your project files easily

These features save time and reduce frustration, especially when learning.

## Step 1: Download VS Code

### Go to the Official Website

1. Open your web browser
2. Go to: **[https://code.visualstudio.com/](https://code.visualstudio.com/)**
3. Click the large **Download** button

The website automatically detects your operating system (Windows, Mac, or Linux) and offers the correct version.

```{tip}
If you need a different version (e.g., 32-bit Windows or a specific Linux distribution), click the small dropdown arrow next to the download button.
```

## Step 2: Install VS Code

### Windows

1. **Find the downloaded file** â€” Usually in your `Downloads` folder, named something like `VSCodeUserSetup-x64-1.xx.x.exe`
2. **Double-click** to run the installer
3. **Accept the license agreement** and click Next
4. **Choose installation location** â€” The default is fine for most users
5. **Select additional tasks** (recommended):
   - âœ… Add "Open with Code" action to file context menu
   - âœ… Add "Open with Code" action to directory context menu
   - âœ… Add to PATH (important!)
6. Click **Install** and wait for completion
7. Click **Finish** (optionally launch VS Code)

### macOS

1. **Find the downloaded file** â€” Usually in your `Downloads` folder, named `VSCode-darwin-universal.zip`
2. **Double-click** to extract the application
3. **Drag** `Visual Studio Code.app` to your **Applications** folder
4. **Open** VS Code from Applications (you may need to right-click â†’ Open the first time due to security settings)

```{tip}
**macOS security warning:** If you see "VS Code can't be opened because it is from an unidentified developer", go to System Preferences â†’ Security & Privacy â†’ General, and click "Open Anyway".
```

### Linux

The easiest method depends on your distribution:

**Ubuntu/Debian:**
```bash
# Download and install the .deb package from the website, then:
sudo dpkg -i code_*.deb
```

**Or use snap:**
```bash
sudo snap install code --classic
```

See the [official Linux installation guide](https://code.visualstudio.com/docs/setup/linux) for other distributions.
