## Why Do We Need a Python Extension?

VS Code is a general-purpose text editorâ€”it doesn't understand Python by default. It's like having a word processor that can display text but doesn't know about spelling or grammar.

The **Python extension** teaches VS Code to understand Python code, giving you:

| Feature | What It Does | Why It Helps |
|---------|--------------|---------------|
| **Syntax highlighting** | Colors different parts of code | Easier to read and spot errors |
| **IntelliSense** | Suggests code as you type | Faster coding, fewer typos |
| **Error detection** | Underlines problems in red | Find bugs before running code |
| **Code navigation** | Jump to function definitions | Understand code faster |
| **Debugging** | Step through code line by line | Find and fix problems |
| **Jupyter support** | Run notebooks in VS Code | Interactive data analysis |

```
import pandas as pd        â† All text looks the same
df = pd.read_csv('data')   â† No suggestions while typing
print(df.head()            â† Missing parenthesis not detected
```

**With the Python extension:**
```python
import pandas as pd        â† Keywords colored differently
df = pd.read_csv('data')   â† Autocomplete suggests methods
print(df.head()            â† Red underline shows the error!
```

## Step 2: Install the Python Extension

1. In the search bar, type: **Python**
2. Look for **"Python"** by **Microsoft** (it should be the first result)
   - It has millions of downloads
   - The icon is a blue and yellow Python logo
3. Click the **Install** button
4. Wait for installation to complete (usually just a few seconds)

```{warning}
There are several Python-related extensions. Make sure you install the one simply called **"Python"** by **Microsoft**, not "Python Extension Pack" or other variants. The Microsoft one is the official extension.
```

### What Gets Installed

The Python extension automatically includes:
- **Pylance** â€” Fast, feature-rich language support
- **Python Debugger** â€” For stepping through code

You don't need to install these separately.

## Step 3: Verify the Installation

Let's confirm the extension is working.

### Create a Test Python File

1. Go to **File â†’ New File** (or press `Ctrl+N`)
2. Go to **File â†’ Save As** (or press `Ctrl+S`)
3. Save the file as `test.py` (the `.py` extension is important!)
4. Choose any location (your Desktop is fine for now)

### Test Syntax Highlighting

Type or paste this code into your `test.py` file:

```python
# This is a comment
message = "Hello, Water Modeller!"
number = 42
print(message)
```

**What you should see:**
- `# This is a comment` â€” appears in a different color (often green or gray)
- `"Hello, Water Modeller!"` â€” the string appears in another color (often orange or brown)
- `print` â€” the function name is highlighted (often yellow or blue)
- `42` â€” numbers may have their own color

If you see different colors for different parts of the code, **syntax highlighting is working!**

### Test IntelliSense (Autocomplete)

1. On a new line, type: `pri`

If suggestions appear as you type, **IntelliSense is working!**

## Troubleshooting

### "I don't see any syntax highlighting"

- Make sure your file is saved with a `.py` extension
- Check the bottom-right corner of VS Codeâ€”it should say "Python"
- If it says "Plain Text", click it and select "Python"

### "The extension won't install"

- Check your internet connection
- Try restarting VS Code
- On work computers, your IT department may need to allow VS Code marketplace access

### "I see an error about Python not being installed"

This is expected! We haven't installed Python yet. The extension is working correctlyâ€”it's just telling you that Python itself isn't set up. We'll fix this in Module 2b when we install `uv`.

### "IntelliSense suggestions are slow or don't appear"

- Wait a few seconds after opening a fileâ€”the extension needs time to analyze your code
- Make sure Pylance is enabled (it's included with the Python extension)
- Try pressing `Ctrl+Space` to manually trigger suggestions
